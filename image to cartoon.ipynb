# Convert image into cartoonish image

```python
import cv2
import numpy as np
```

```python
img=cv2.imread("DIVYA.jpg")
img
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
grey
```

```python
plt.imshow(grey)
```

```python
grey=cv2.medianBlur(grey,9)
grey
```

```python
plt.imshow(grey)
```

```python
edges=cv2.adaptiveThreshold(grey,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,13,4)
edges
```

```python
plt.imshow(edges)
```

```python
color=cv2.bilateralFilter(img,15,350,555)
color
```

```python
plt.imshow(color)
```

```python
cartoon=cv2.bitwise_and(color,color,mask=edges)
cartoon
```

```python
plt.imshow(cartoon)
```

```python
sho=cv2.imshow("show",cartoon)
sho
```
